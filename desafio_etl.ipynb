```python
# Lê o CSV e converte para uma lista de dicionários
usuarios = pd.read_csv('lista.csv').to_dict(orient='records')

# Garante a estrutura esperada para a etapa de Transformação
for usuario in usuarios:
    usuario['news'] = []
```

Etapa de Extração

```python
import pandas as pd

df = pd.read_csv('lista.csv')
usuario_ids = df['id'].tolist()
print(usuario_ids)
```

```python
import pandas as pd

df = pd.read_csv('lista.csv')
usuario_receitas = df['receita'].tolist()
print(usuario_receitas)
```

```python
import pandas as pd

df = pd.read_csv('lista.csv')
usuario_despesas = df['despesa'].tolist()
print(usuario_despesas)
```

Etapa de Transformação (Calculando se consegue pagar ou não)

```python
def condicao(usuario):
  lucro = usuario_receitas[usuario] - usuario_despesas[usuario]
  return "Em crédito OK" if lucro >= 0 else "Em débito X"

repetidor = 0
usuario_lucros = []
for usuario in usuarios:
  pegando_lucros = condicao(repetidor)
  usuario_lucros.append(pegando_lucros)
  repetidor = repetidor + 1
  print(pegando_lucros)
  print(usuario_lucros)

# vezes = 0
# for usuario in usuarios:
#   news = condicao(vezes)
#   usuario['news'] = news
#   vezes = vezes + 1
#   print(news)
#   print(usuario)
```

Etapa de Carregamento (Salvar dados transformados no arquivo)

```python
final_repetidor = 0
for usuario in usuarios:
  news = usuario_lucros[final_repetidor]
  usuario['news'] = news
  print(f"Usuário do Id {usuario_ids[final_repetidor]} foi atualizado")
  print(usuario)
  final_repetidor = final_repetidor + 1

df['status'] = usuario_lucros

df.to_csv("lista.csv", index=False)
```
